## RQ2-noise

```python
import pandas as pd
import os

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth', 200)


path_list = ['results/apfd_accident_C3D_10.csv',
             'results/apfd_accident_R2Plus1D_50.csv',
             'results/apfd_accident_R3D_10.csv',
             'results/apfd_accident_slowfastnet_40.csv',
             'results/apfd_hmdb51_C3D_40.csv',
             'results/apfd_hmdb51_R2Plus1D_90.csv',
             'results/apfd_hmdb51_R3D_40.csv',
             'results/apfd_hmdb51_slowfastnet_40.csv',
             'results/apfd_ucf101_C3D_18.csv',
             'results/apfd_ucf101_R2Plus1D_21.csv',
             'results/apfd_ucf101_R3D_10.csv',
             'results/apfd_ucf101_slowfastnet_24.csv']

data_list = ['Accident', 'Accident', 'Accident', 'Accident',
             'HMDB51', 'HMDB51', 'HMDB51', 'HMDB51',
             'UCF101', 'UCF101', 'UCF101', 'UCF101']

model_list = ['C3D', 'R2Plus1D', 'R3D', 'SlowFastNet', 
              'C3D', 'R2Plus1D', 'R3D', 'SlowFastNet', 
              'C3D', 'R2Plus1D', 'R3D', 'SlowFastNet']
approach_list = ['XGB', 'LGB', 'RF', 'LR', 'DeepGini', 'VanillaSM', 'PCS', 'Entropy', 'Random']

noise_list = ['channel_shift_range', 'featurewise_std_normalization', 'height_shift', 'horizontal_flip',
             'shear_range', 'width_shift', 'zca_whitening']

def get_noise_re(path, model_name):
    df = pd.read_csv(path)
    del df['name']
    re = df.to_numpy()[1:, :]
    df = pd.DataFrame(re, columns=approach_list)
    df['Noise'] = noise_list
    df['Model'] = model_name
    return df


df_list = []
for i in path_list:
    df_list.append(get_noise_re(i, i.split('/')[-1].split('_')[2]))
df = pd.concat(df_list, ignore_index=True)
df = df[['Noise', 'Model', 'Random', 'DeepGini', 'VanillaSM', 'PCS', 'Entropy', 'LGB']]
df = df.groupby(['Noise', 'Model']).agg('mean').reset_index()

noise_list = ['horizontal_flip',
             'height_shift',
             'width_shift',
             'featurewise_std_normalization',
             'shear_range',
             'zca_whitening',
             'channel_shift_range']

approach_list = ['Random', 'DeepGini', 'VanillaSM', 'PCS', 'Entropy', 'VRank']
model_list = ['C3D', 'R2Plus1D', 'R3D', 'slowfastnet']

df_list = []
for noise in noise_list:
    m = df[df['Noise']==noise].T.to_numpy()[2:,:]
    tmp_df = pd.DataFrame(m, columns=['C3D', 'R2Plus1D', 'R3D', 'slowfastnet'])
    tmp_df['Noise'] = noise
    tmp_df['Approach'] = approach_list
    df_list.append(tmp_df)
df = pd.concat(df_list, ignore_index=True)
df = df[['Noise','Approach', 'C3D', 'R2Plus1D', 'R3D', 'slowfastnet']]
df['C3D'] = [str(round(i, 3)) for i in df['C3D']]
df['R2Plus1D'] = [str(round(i, 3)) for i in df['R2Plus1D']]
df['R3D'] = [str(round(i, 3)) for i in df['R3D']]
df['slowfastnet'] = [str(round(i, 3)) for i in df['slowfastnet']]
df_re = df.copy()
# df_re.to_excel("tables/noise_res.xlsx",index=True)
print(df_re.head(10))
res = df_re.to_numpy()

del df_re['Noise']
del df_re['Approach']
m = df_re.to_numpy().astype(float)

improve_res = []
for i in range(len(m)):
    if 0<=i<=5:
        tmp = (m[5]-m[i]) / m[i]
        improve_res.append(tmp)
    elif 6<=i<=11:
        tmp = (m[11]-m[i]) / m[i]
        improve_res.append(tmp)
    elif 12<=i<=17:
        tmp = (m[17]-m[i]) / m[i]
        improve_res.append(tmp)
    elif 18<=i<=23:
        tmp = (m[23]-m[i]) / m[i]
        improve_res.append(tmp)
    elif 24<=i<=29:
        tmp = (m[29]-m[i]) / m[i]
        improve_res.append(tmp)
    elif 30<=i<=35:
        tmp = (m[35]-m[i]) / m[i]
        improve_res.append(tmp)
    elif 36<=i<=41:
        tmp = (m[41]-m[i]) / m[i]
        improve_res.append(tmp)
        
def get_4_value(i):
    i = "%.2f%%" % (i*100)
    i = i +'0'* (5-len(i))
    return i

new_res = []
for i in range(len(improve_res)):
    tmp_new_res = []
    for j in range(len(improve_res[i])):
        if improve_res[i][j] == 0:
            tmp_new_res.append('-')
        else:
            tmp_new_res.append(get_4_value(improve_res[i][j]))
    new_res.append(tmp_new_res)
    
new_res = np.array(new_res)

all_res = np.concatenate((res, new_res), axis=1)

df_rq2 = pd.DataFrame(all_res, columns=['Noise', 'Approach']+['C3D', 'R2Plus1D', 'R3D', 'slowfastnet']*2)

# df_rq2.to_excel('tables/noise_res.xlsx', index=False)

df_rq2
```
